```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import bokeh.io
import bokeh.mpl
import bokeh.plotting

import pandas_datareader.data as web
import datetime

from bokeh.plotting import figure, show
from bokeh.charts import Bar, Horizon, output_file, show

import scipy
```

```python
# set inline viewing
bokeh.io.output_notebook()
```

```python
# load data
start = datetime.datetime(2017, 1, 1)
end = datetime.datetime(2017, 6, 29)
```

```python
tsla = web.DataReader("BRK.B", 'google', start, end)
tsla = tsla.drop(tsla.columns[[0, 1, 2, 4]], axis=1)
#tsla = tsla.rename(columns={'Close':'TSLA'})
tsla['Returns']=tsla['Close'].pct_change(1).cumsum()
aapl = web.DataReader("AAPL", 'google', start, end)
aapl = aapl.drop(aapl.columns[[0, 1, 2, 4]], axis=1) 
#aapl = aapl.rename(columns={'Close':'AAPL'})
aapl['Returns']=aapl['Close'].pct_change(1).cumsum()
```

```python
from scipy.stats import linregress
x = tsla['Close']
y = aapl['Close']
slope, intercept, r_value, p_value, slope_std_error = linregress(x, y)

# Calculate some additional outputs
predict_y = intercept + slope * x.reset_index().index
#pred_error = y - predict_y
#degrees_of_freedom = len(x) - 2
#residual_std_error = np.sqrt(np.sum(pred_error**2) / degrees_of_freedom)

# Plotting
#pylab.plot(x, y, 'o')
#pylab.plot(x, predict_y, 'k-')
#pylab.show()
```

```python
#p = Bar(tsla.reset_index(), 'Close', values='Date')
#show(p)
p = figure(title='TSLA vs AAPL', x_axis_type='datetime',x_axis_label='Date', y_axis_label='Close')
p.line(tsla.index,tsla['Close'],color='red')
#p.vbar(tsla.index,top=tsla['Volume'],width=0.5,color='red')
p.line(aapl.index,aapl['Close'],color='blue')
#p.circle(aapl.index, aapl['Close'], legend="y=x")
p.line(tsla.index,predict_y)
show(p)
```

```python
p = figure(title='TSLA vs AAPL', x_axis_type='datetime',x_axis_label='Date', y_axis_label='Close')
p.line(tsla.index,tsla['Returns'],color='red')
#p.vbar(tsla.index,top=tsla['Volume'],width=0.5,color='red')
p.line(aapl.index,aapl['Returns'],color='blue')
#p.circle(aapl.index, aapl['Close'], legend="y=x")
show(p)
```
